```python
list_games = []
for number in range(30):
    random.shuffle(deck_list)
    length_to_split = [13, 13, 13, 13]
    Inputt = iter(deck_list)
    Output = [list(islice(Inputt, elem))
    for elem in length_to_split]
    list_games.append(Output)
len(list_games)
```
